```python
from bs4 import BeautifulSoup
import requests 
url = 'https://www.merco.info/pe/ranking-merco-empresas'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[0]
```

```python
table = soup.find_all('table')[0]
print(table)
```

```python
titulos = table.find_all('th')
```

```python
titulos_tabla = [cabecera.text.strip() for cabecera in titulos]
print(titulos_tabla)
```

```python
## ANOTHER OPTION:
## from collections import OrderedDict
## titulos_tabla_unicos = list(OrderedDict.fromkeys(titulos_tabla))
## print(titulos_tabla_unicos)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = titulos_tabla)
```

```python
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = (row.find_all('td'))
    individual_row_data = [data.text.strip() for data in row_data]
        
    length = len(df)
    df.loc[length] = individual_row_data
```
